```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from ipywidgets import interact, FloatSlider

def generar_senal_compuesta(frecuencias, duracion=1.0, fs=1000):
    """
    Genera una seal compuesta por mltiples frecuencias senoidales.

    Args:
        frecuencias (list): Lista de frecuencias en Hz
        duracion (float): Duracin de la seal en segundos
        fs (int): Frecuencia de muestreo en Hz

    Returns:
        t (np.array): Array de tiempos
        senal (np.array): Seal compuesta
    """
    t = np.linspace(0, duracion, int(fs * duracion), endpoint=False)
    senal = np.zeros_like(t)

    for freq in frecuencias:
        senal += np.sin(2 * np.pi * freq * t)

    # Normalizamos la seal para que tenga amplitud mxima 1
    senal = senal / np.max(np.abs(senal))

    return t, senal

def analizar_convolucion(frecuencia_test):
    # Frecuencias de la seal compuesta
    frecuencias = [5, 20, 50, 100]  # Hz

    # Generar seal compuesta
    t, senal_compuesta = generar_senal_compuesta(frecuencias)

    # Generar seal de prueba (senoidal)
    senal_test = np.sin(2 * np.pi * frecuencia_test * t)

    # Calcular convolucin
    convolucion = signal.convolve(senal_compuesta, senal_test, mode='same')

    # Calcular valor mximo de la convolucin
    max_convolucion = np.max(np.abs(convolucion))

    # Crear figura
    plt.figure(figsize=(12, 8))

    # Seal compuesta
    plt.subplot(3, 1, 1)
    plt.plot(t, senal_compuesta)
    plt.title(f'Seal compuesta (Frecuencias: {frecuencias} Hz)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    # Seal de prueba
    plt.subplot(3, 1, 2)
    plt.plot(t, senal_test)
    plt.title(f'Seal de prueba (Frecuencia: {frecuencia_test} Hz)')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    # Convolucin
    plt.subplot(3, 1, 3)
    plt.plot(t, convolucion)
    plt.title(f'Convolucin (Mximo: {max_convolucion:.2f})')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Mostrar frecuencias donde la convolucin es mxima
    print(f"La convolucin alcanza su mximo cuando la frecuencia de prueba coincide con una de las frecuencias de la seal compuesta.")
    print(f"Frecuencias en seal compuesta: {frecuencias} Hz")

# Crear interfaz interactiva
interact(analizar_convolucion,
         frecuencia_test=FloatSlider(min=1, max=150, step=1, value=20,
                                    description='Frecuencia prueba (Hz)'))
```
