```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Configuración de parámetros
fs = 1000  # Frecuencia de muestreo (Hz)
T = 10     # Duración de la señal (segundos)
t = np.linspace(0, T, int(fs*T), endpoint=False)  # Vector de tiempo

# Crear señal chirp (frecuencia aumenta con el tiempo)
f0 = 1    # Frecuencia inicial (Hz)
f1 = 100  # Frecuencia final (Hz)
signal = np.sin(2*np.pi*(f0 + (f1-f0)*t/(2*T))*t)

# Calcular la Transformada de Fourier
n = len(signal)
yf = fft(signal)
xf = fftfreq(n, 1/fs)[:n//2]

# Graficar la señal en el tiempo y su transformada de Fourier
plt.figure(figsize=(12, 6))

# Gráfico de la señal en el tiempo
plt.subplot(1, 2, 1)
plt.plot(t, signal)
plt.title('Señal Chirp en el Dominio del Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)

# Gráfico de la transformada de Fourier
plt.subplot(1, 2, 2)
plt.plot(xf, 2/n * np.abs(yf[0:n//2]))
plt.title('Transformada de Fourier')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim(0, 120)  # Limitamos el eje x para mejor visualización

plt.tight_layout()
plt.show()
```
